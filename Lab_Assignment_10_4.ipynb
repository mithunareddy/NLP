```python
print("Loading Word2Vec model...")
model = api.load("word2vec-google-news-300")

print("Vocabulary size:", len(model.key_to_index))

# show one example vector
word = "computer"
print("Example word:", word)
print("Vector length:", len(model[word]))
print("Vector:", model[word])
```

```python
words = [
    # Animals
    "dog","cat","lion","tiger","elephant","horse","cow","monkey",
    # Cities
    "india","london","paris","tokyo","delhi","mumbai","chennai","newyork",
    # Technology
    "computer","laptop","mobile","internet","software","hardware","keyboard","mouse",
    # Emotions
    "happy","sad","angry","fear","love","joy","smile","cry",
    # Vehicles
    "car","bus","train","airplane","bike","ship","truck","cycle"
]
print("Total words selected:", len(words))
```

```python
vectors = []

for word in words:
    vectors.append(model[word])

vectors = np.array(vectors)
print("Vector shape:", vectors.shape)
```

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=10)
reduced_vectors = tsne.fit_transform(vectors)
print("Reduced shape:", reduced_vectors.shape)
```

```python
plt.figure(figsize=(12,10))
x = reduced_vectors[:,0]
y = reduced_vectors[:,1]
plt.scatter(x, y)
for i, word in enumerate(words):
    plt.annotate(word, (x[i], y[i]))
plt.title("t-SNE Visualization of Word Embeddings")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.grid(True)
plt.show()
```
