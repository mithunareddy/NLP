```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = "Loving the new AI features üòç 2025 #AI #MachineLearning"
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```

# **TASK #3**

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

essay_text = """
Academic writing plays a crucial role in the dissemination of knowledge.
Researchers analyze data, develop theories, and present arguments in a structured manner.
The study demonstrates how linguistic patterns influence clarity and coherence.
Effective communication enhances understanding and supports critical evaluation.
"""

essay_text_word_tokens = nltk.word_tokenize(essay_text)
essay_text_word_tokens_pos_tags = nltk.pos_tag(essay_text_word_tokens)
print(essay_text_word_tokens_pos_tags)
```

```python
import spacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")

essay_text = """
Academic writing plays a crucial role in the dissemination of knowledge.
Researchers analyze data, develop theories, and present arguments in a structured manner.
The study demonstrates how linguistic patterns influence clarity and coherence.
Effective communication enhances understanding and supports critical evaluation.
"""
doc = nlp(essay_text)

noun_words = []
verb_words = []

for token in doc:
  if token.pos_ in ["NOUN", "PROPN"]:
    noun_words.append(token.text)
  elif token.pos_ == "VERB":
    verb_words.append(token.text)

noun_freq = Counter(noun_words)
verb_freq = Counter(verb_words)

print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```
